```python
# import all important package
import matplotlib.pyplot as plt
from matplotlib.figure import projections
from scipy.io import netcdf
import cartopy.crs as ccrs
import cartopy.feature
```

```python
# Read data
dataset = netcdf.netcdf_file('data_project/SeaSurfaceTemp.nc',
                             maskandscale=True, mmap=False)
dataset
```

```python
# Create new variable to define the data
df_sst = dataset.variables['SST'][0, :, :]
df_lat = dataset.variables['LAT'][:]
df_lon = dataset.variables['LONN719_721'][:]
```

```python
# Visualization of Sea Surface Temperature
ax = plt.axes(projection=ccrs.Mollweide())
plt.contourf(df_lon, df_lat, df_sst,transform=ccrs.PlateCarree())
plt.show()
```

```python
ax = plt.axes(projection=ccrs.Mollweide())
plot = plt.contourf(df_lon, df_lat, df_sst, transform=ccrs.PlateCarree())
plt.colorbar(plot, ax=ax, shrink=0.4)
plt.show()
```
